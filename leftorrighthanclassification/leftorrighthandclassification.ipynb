```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

import tensorflow as tf
```

```python
df = pd.read_csv("otu.csv")
```

```python
df.head()
```

```python
dft = df.transpose()
dft
```

```python
features = dft[dft.columns[1:]].values
expectedOutputs = dft[dft.columns[0]].values;
```

```python
features = features.astype(np.double)
features
```

```python
expectedOutputs[expectedOutputs == 'left'] = 0
expectedOutputs[expectedOutputs == 'right'] = 1
expectedOutputs = expectedOutputs.astype(np.double)
expectedOutputs
```

```python
means = np.sum(features, axis=0) / np.size(features, 1)
reducedFeatures = np.delete(features, means == 0, 1)

reducedFeatures.shape
```

```python
scaler = StandardScaler()
reducedFeatures = scaler.fit_transform(reducedFeatures)
```

```python
featureTrain, featureTest, expectedTrain, expectedTest = train_test_split(reducedFeatures, expectedOutputs, test_size=0.2, random_state=7)
#featureTrain, featureTemp, expectedTrain, expectedTemp = train_test_split(features, expectedOutputs, test_size=0.4, random_state=12)
#featureValid, featureTest, expectedValid, expectedTest = train_test_split(featureTemp, expectedTemp, test_size=0.5, random_state=7)
```

```python
expectedTrain
```

```python
tf.convert_to_tensor(featureTrain)
tf.convert_to_tensor(expectedTrain)
tf.convert_to_tensor(featureTest)
tf.convert_to_tensor(expectedTest)
expectedTrain
```

```python
model1 = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
    ])
```

```python
model1.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.7),
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics=['accuracy', 'AUC', tf.keras.metrics.TruePositives(), 
              tf.keras.metrics.TrueNegatives(), tf.keras.metrics.FalsePositives(),
              tf.keras.metrics.FalseNegatives()])
```

```python
model1.evaluate(featureTrain, expectedTrain)
```

```python
model1.fit(featureTrain, expectedTrain, batch_size=32, epochs=50)
#model.fit(featureTrain, expectedTrain, batch_size=16, epochs=50, validation_data=(featureValid, expectedValid))
```

```python
mets = model1.evaluate(featureTest, expectedTest)
auc = mets[2];
tp = mets[3];
tn = mets[4];
fp = mets[5];
fn = mets[6];

sensitivity = tp / (tp + fn)
specificity = tn / (tn + fp)

print(f'AUC: {auc}')
print(f'sensitivity: {sensitivity}')
print(f'specificity: {specificity}')
```
